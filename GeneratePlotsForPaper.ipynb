```python
import pandas as pd
import matplotlib.pyplot as plt

# filename = 'stat_train_9792_2.csv'
filename_heuristic = 'results\heuristic_test_surge_run47_revised.csv'
filename_model_1 = 'results\maddpg_shariq_test_surge_run73_revised.csv'
filename_model_2 = 'results\static_test_surge_run74_revised.csv'
filename_model_3 = 'results\maddpg_shariq_test_surge_LowAfternoon_run74_revised.csv'
df_heuristic = pd.read_csv(filename_heuristic)
df_model_1 = pd.read_csv(filename_model_1)
df_model_2 = pd.read_csv(filename_model_2)
df_model_3 = pd.read_csv(filename_model_3)

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df_heuristic["avg_waiting_time_car"]
x1 = df_model_1["avg_waiting_time_car"]
x2 = df_model_2["avg_waiting_time_car"]

plt.plot(x, label = "Heuristic")
plt.plot(x1, label = "M_1200 episodes")
plt.plot(x2, label = "Static")
plt.xlabel("30 mins time slot of a day")
plt.ylabel("avg_waiting_time_car")
plt.legend()
```
