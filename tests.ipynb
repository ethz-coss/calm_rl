```python
import pandas as pd
import matplotlib.pyplot as plt

# filename = 'stat_train_9792_2.csv'
filename = 'stat_train_Run71.csv'
_df = pd.read_csv(filename)

df = _df.tail(500)
df.columns
len(df)
```

```python
import seaborn as sns

xvar = 'Car_Flow_Rate'
yvar = 'Car_Lane_Width'
df['color'] = df.apply(lambda x: 'r' if x.RewardAgent_2<0 else 'g', axis=1)
plt.scatter(df[xvar], df[yvar], color=df['color'], s=0.5, alpha=0.3)
```

```python
xvar = 'Ped_Flow_Rate'
yvar = 'Ped_Lane_Width'
df['color'] = df.apply(lambda x: 'r' if x.RewardAgent_2<0 else 'g', axis=1)
sns.scatterplot(data=df, x=xvar, y=yvar, hue="Co_Sharing")#, color=df['color'], s=0.5, alpha=0.3)
```
