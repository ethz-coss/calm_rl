```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

mapping = {
    # 'continuous_augmented': 'model_test_continuous_augmented',
    #        'discrete': 'model_test_discrete',
    #        'continuous_load': 'model_test_continuous',
        #    'continue_full_reload': 'model_better_continuous',
        #    'heuristic': 'model_heuristic',
        #    'tweaked_reward': 'model_reward_coshare_0.1',
            # 'back2basics': 'model_back2basics',
            # 'gumbel_cont': 'model_gumbel_continuous',
            # 'fixed_coshare': 'model_test_surge_run47',
            'noBatchNormShariq': 'results/static_test_surge_92noBatchNorm',
            'mitBatchNormShariq': 'results/static_test_surge_run92',
            # 'rohit': 'results/maddpg_shariq_test_surge_run73'
           }


keys_to_plot = [
    'cosharing', 'avg_waiting_time_ped', 'los',
    'avg_queue_length_car', 'avg_queue_length_bike', 'avg_queue_length_ped'
]

fig, axes = plt.subplots(2,3, figsize=(10,8), dpi=300)
axes = axes.flatten()
for i, (label, _filename) in enumerate(mapping.items()):
    filename = f'{_filename}.csv'
    df = pd.read_csv(filename)
    df['cosharing'] *= 1
    print(label, np.mean(df['Reward_Agent_2']>0))

    agg = df.groupby('timeslot').mean()
    for j, key in enumerate(keys_to_plot):
        axes[j].plot(agg.index, agg[key], label=label)
        if i==0:
            axes[j].set_title(key)
axes[2].legend()
```
